```python
rede =  local_dir+'/rede_palavra_chave_x_periodico.xlsx'
df_rede.to_excel(rede,sheet_name='Sheet1',header=True, index=False)
df_rede.head()
```

```python
colunas = []
for coluna in list(df_rede):
    colunas.append(coluna+' varchar\n')  
sql_create_table = 'drop table if exists alacip.rede_palavra_chave_x_periodico;\ncreate table alacip.rede_palavra_chave_x_periodico (\n'
sql_create_table = sql_create_table + ','.join(colunas) + ');\n'

try:
    conn = pg.connect(connect_cmd)
except:
    print ("I am unable to connect to the database")

cur = conn.cursor()
try:
    cur.execute(sql_create_table)
    conn.commit()
except:
    print("erro")
    conn.rollback()
conn.close()

print('a tabela  alacip.rede_palavra_chave_x_periodico foi criada')
```

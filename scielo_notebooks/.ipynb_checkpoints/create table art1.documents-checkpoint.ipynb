```python
query_create_table_documents = """
drop table if exists art1.documents ;
create table art1.documents (
-- authors 
         document_extraction_date varchar,
         document_study_unit varchar,
         document_collection varchar,
         document_issn_scielo varchar,
         document_issns varchar,
         document_title_at_scielo varchar,
         document_title_thematic_areas varchar,
         document_title_is_agricultural_sciences varchar,
         document_title_is_applied_social_sciences varchar,
         document_title_is_biological_sciences varchar,
         document_title_is_engineering varchar,
         document_title_is_exact_and_earth_sciences varchar,
         document_title_is_health_sciences varchar,
         document_title_is_human_sciences varchar,
         document_title_is_linguistics_letters_and_arts varchar,
         document_title_is_multidisciplinary varchar,
         document_title_current_status varchar,
         document_publishing_id_scielo varchar,
         document_publishing_year varchar,
         document_type varchar,
         document_is_citable varchar,
         document_author varchar,
         document_author_institution varchar,
         document_author_affiliation_country varchar,
         document_author_affiliation_state varchar,
         document_author_affiliation_city varchar,
-- affilliation  
         document_affiliation_instituition varchar,
         document_affiliation_country varchar,
         document_affiliation_country_iso_3166 varchar,
         document_affiliation_state varchar,
         document_affiliation_city varchar,
-- dates  
         document_submitted_at varchar,
         document_submitted_at_year varchar,
         document_submitted_at_month varchar,
         document_submitted_at_day varchar,
         document_accepted_at varchar,
         document_accepted_at_year varchar,
         document_accepted_at_month varchar,
         document_accepted_at_day varchar,
         document_reviewed_at varchar,
         document_reviewed_at_year varchar,
         document_reviewed_at_month varchar,
         document_reviewed_at_day varchar,
         document_published_as_ahead_of_print_at varchar,
         document_published_as_ahead_of_print_at_year varchar,
         document_published_as_ahead_of_print_at_month varchar,
         document_published_as_ahead_of_print_at_day varchar,
         document_published_at varchar,
         document_published_at_year varchar,
         document_published_at_month varchar,
         document_published_at_day varchar,
         document_published_in_scielo_at varchar,
         document_published_in_scielo_at_year varchar,
         document_published_in_scielo_at_month varchar,
         document_published_in_scielo_at_day varchar,
         document_updated_in_scielo_at varchar,
         document_updated_in_scielo_at_year varchar,
         document_updated_in_scielo_at_month varchar,
         document_updated_in_scielo_at_day varchar,
-- languages 
         document_languages varchar,
         document_pt varchar,
         document_es varchar,
         document_en varchar,
         document_other_languages varchar,
-- licence   
         document_license varchar,
-- altimetric                  
         document_altimetric_score varchar,
         document_altmetrics_url varchar
);

create index on art1.documents(document_issn_scielo)
"""

import psycopg2 as pg
try:
    conn = pg.connect(connect_cmd)
except:
    print ("I am unable to connect to the database")

cur = conn.cursor()
try:
    cur.execute(query_create_table_documents)
    conn.commit()
    print('a tabela domuments foi criada')
except:
    print("erro")
    conn.rollback()
conn.close()
```

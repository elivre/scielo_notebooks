```python
query_artigos_sem_autor = """
 select scielo_id,titulo,'http://www.scielo.br/scielo.php?script=sci_arttext&pid='||scielo_id||'&lang=pt' as link_html
 from alacip.artigos_selecionados
 where autor_1_nome is null 
"""
conn=pg.connect(connect_cmd)
cur = conn.cursor()
cur.execute(query_artigos_sem_autor)
for id in cur:
    print('Titulo: '+id[1])
    print('http://www.scielo.br/scielo.php?script=sci_arttext&pid='+id[0]+'&lang=pt')
df=pd.read_sql(query_artigos_sem_autor,conn)
writer = pd.ExcelWriter(local_dir+'/artigos_sem_autor.xlsx')
df.to_excel(writer, sheet_name='Sheet1',header=True, index=False)
```

```python

query_artigos_XML_Invalido = """
 select scielo_id,titulo,'http://www.scielo.br/scielo.php?script=sci_arttext&pid='||scielo_id||'&lang=pt' as link_html
 from alacip.xml_dados
 where texto like 'XML Invalido' 
"""
conn=pg.connect(connect_cmd)
cur = conn.cursor()
cur.execute(query_artigos_sem_autor)
for id in cur:
    print('Titulo: '+id[1])
    print('http://www.scielo.br/scielo.php?script=sci_arttext&pid='+id[0]+'&lang=pt')
df=pd.read_sql(query_artigos_XML_Invalido,conn)
writer = pd.ExcelWriter(local_dir+'/artigos_com_xml_invalido.xlsx')
df.to_excel(writer, sheet_name='Sheet1',header=True, index=False)
```

```python
import pandas as pd
import numpy as np

query_nome_de_autores_M_ou_F_indefino = """
select split_part(autor_nome_completo, ' ', 1)  as nome, '' as sexo
from alacip.autores_selecionados
where autor_sexo is null
group by nome
order by nome;
"""

conn = pg.connect(connect_cmd)
df_doacoes = pd.read_sql(query_nome_de_autores_M_ou_F_indefino, conn)
df_doacoes.fillna('#NULO')
planilha =  local_dir+'/nomes_M_ou_F.xlsx'
df_doacoes.to_excel(planilha,sheet_name='Sheet1',header=True, index=False)
df_doacoes.head()
conn.close()
print("OK")
```

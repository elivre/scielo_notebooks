```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = 'http://www.scielo.br/scielo.php?script=sci_abstract&pid=S0104-62762016000100028&lng=en&nrm=iso&tlng=pt'
h=requests.get(url).text
soup = BeautifulSoup(h,  'html.parser') 

print("pdf link ---------------------------------------")
p=soup.select('a[href^="/pdf/"]')
print(p)

print("read portuguese pdf link -----------------------")
p=soup.select('a[href$="Portuguese (pdf)"]')
print(p)

print("read citation autors ---------------------------")
r=soup.select('meta[name^="citation_author"]')
for i in r:
    print(i)

#r = requests.get(url)
#data = r.text
#data

```

```python

dados_autores = """
select a.*, 'controle' as journal_class from documents_authors as a
where issn_scielo in (
    -- controle
    '0104-9313',
    '0103-2070',
    '0102-0188',
    '0034-7140',
    '0101-3173',
    '1415-4757'
)
union
select a.*, 'objeto' as journal_class from documents_authors as a
where issn_scielo in (
    -- objeto 
    '0011-5258',
    '0104-6276',
    '1981-3821',
    '0103-3352',
    '0102-6445',
    '0034-7329',
    '0102-8529',
    '0104-4478'
);
"""
%sql insert into art1.autores {dados_autores}


```

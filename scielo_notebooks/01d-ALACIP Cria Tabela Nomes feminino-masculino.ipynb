## cria um arquivo de nomes femininos e masculinos com base nos dados das paginas do site 
https://www.dicionariodenomesproprios.com.br

```python
import urllib
from bs4 import BeautifulSoup

def write_nomes(file,home_page,n_paginas,sexo):
    nomes = open(file,"a")
    for pg in range(1,n_paginas+1):
        url = home_page+"/"+str(pg)+"/"
        print(url)
        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html)

        n = soup.find_all("a","lista-nome","href")

        for l in n:
            nomes.write(l.get_text()+';'+sexo+'\n')
    nomes.close()
 
arq_csv = local_dir+'/nomes_propios.csv'
arq_nomes = open('nomes_MF.csv','w')
arq_nomes.close()
write_nomes(arq_csv,"https://www.dicionariodenomesproprios.com.br/nomes-femininos",152,'F')
write_nomes(arq_csv,"https://www.dicionariodenomesproprios.com.br/nomes-masculinos",194,'M')        
```

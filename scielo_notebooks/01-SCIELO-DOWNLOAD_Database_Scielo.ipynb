```python
import os
import psycopg2 as pg
import os.path


def insert_table_data(table_name,csv_file_path):


    fh = open(csv_file_path, mode='rt')
    header= fh.readline()

    conn = pg.connect(connect_cmd)

    cur = conn.cursor()

    for line in fh:
        line1 = line.replace(',',',')
        line2 = line1.replace(';', ';')
        line3 = line2.replace("'"," ")        
        line4 = line3.replace('"',"'")    

        query = "INSERT INTO {} ".format(table_name)

        query = query + "  values ({});".format(line4)

        #print (query)

        try:
            cur.execute(query)
            #conn.commit()
        except:
            print("erro")
            print(query)
            conn.rollback()
 
    fh.close
    conn.commit()
    conn.close()
    print("OK")
    
```

```python
import os
import fnmatch


def create_table_sql_script(csv_file_path):
    fh = open(csv_file_path, mode='rt')
    header= fh.readline()
    fh.close()

    ##print(header)

    colunas = header.split(',')

    cols = []
    for c in colunas:
        c1 = c.lstrip().rstrip()
        c2 = c1.replace(';', '')
        c3 = c2.replace("'","")        
        c4 = c3.replace('"',"")  
        c5 = c4.replace('(',"") 
        c6 = c5.replace(')',"")
        c7 = c6.replace(' ',"_")
        c8 = c7.replace('\n',"").lower()
        s0 = c8
        s1=s0.replace('á','a')
        s2=s1.replace('é','e')
        s3=s2.replace('í','i')
        s4=s3.replace('ó','o')
        s5=s4.replace('ú','u')
        s6=s5.replace('ü','u')
        s7=s6.replace('ã','a')
        s8=s7.replace('ẽ','e')
        s9=s8.replace('õ','o')
        s10 = s9.replace('c','c')
        cols.append(s10)

    ##print(cols)

    sql_cols = '     '
    cols_len = len(cols)
    for i in range(0,cols_len-1):
        sql_cols  = sql_cols+ cols[i]+'  varchar,'+'\n     '
    sql_cols  =  sql_cols + cols[cols_len-1]+'  varchar'+'\n'

    ##print(sql_cols)

    table_name = os.path.basename(csv_file_path).replace('.csv','')
    sql_create_table = 'create table '+table_name+' (\n'
    sql_create_table = sql_create_table + sql_cols + ');'

    print('-------------------------------------------')
    print(sql_create_table)
    print('-------------------------------------------\n')

for root, dir, fnames in os.walk(local_dir):
        for fname in fnmatch.filter(fnames, "*.csv"):
            csv_file_path = root+'/' +fname
            create_table_sql_script(csv_file_path)
```

```python
query_empirico_teoricos = """
select 
    substring(scielo_id from 1 for 10) as scielo_id ,
    periodico_sigla as sigla,
    periodico_classe as grupo,
    -- sum(cast(empirico as int) ) as empirico,
    sum(cast(tem_tabela as int )) as tabela ,
    sum(cast(tem_quadro as int)) as quadro ,
    sum(cast(tem_grafico as int)) as grafico ,
    sum(cast(tem_diagrama as int)) as diagrama ,
    sum(cast(tem_figura as int)) as figura,
    count(*) as num_artigos
    from
    alacip.artigos_selecionados
    --where empirico <>'0'
    group by 
    substring(scielo_id from 1 for 10) ,
    periodico_sigla ,
    periodico_classe
    order by grupo, tabela desc
    
    ;
"""
conn = pg.connect(connect_cmd)

df = pd.read_sql(query_empirico_teoricos,conn)
df
```

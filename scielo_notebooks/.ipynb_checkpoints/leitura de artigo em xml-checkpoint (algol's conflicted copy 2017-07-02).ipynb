```python
from lxml import etree

tree = etree.parse(url_xml)

elem = {}
def json_line(tag,path):
    e = tree.xpath(path)
    if not e :
        elem[tag]='None'
    else:   
        elem[tag]=e[0].text

def json_line_attrib(tag,path):
    elem[tag]=tree.xpath(path)[0][0].text


json_line('journal_id','/article/front/journal-meta/journal-id')

json_line('journal_title','/article/front/journal-meta/journal-title')

json_line('abbrev_journal_title','/article/front/journal-meta/abbrev-journal-title')

json_line('issn','/article/front/journal-meta/issn')

json_line('publisher_name','/article/front/journal-meta/publisher/publisher-name')

json_line('article-id','/article/front/article-meta/article-id')

json_line('article-id_doi','/article/front/article-meta/article-id[@pub-id-type="doi"]')

json_line('article-title_pt','/article/front/article-meta/title-group/article-title[@xml:lang="pt"]')

json_line('article-title_en','/article/front/article-meta/title-group/article-title[@xml:lang="en"]')

json_line('article-title_es','/article/front/article-meta/title-group/article-title[@xml:lang="es"]')

json_line('article-title_fr','/article/front/article-meta/title-group/article-title[@xml:lang="fr"]')


r=tree.xpath('/article/front/article-meta/contrib-group/contrib[@contrib-type="author"]/name')
#print(r[0].tag)
#print(r[0].attrib)
for n in range(len(r)):
    surname = 'autor_%s_surname'%(n+1)
    name = 'autor_%s_name'%(n+1)
    elem[surname] = r[n][0].text
    elem[name] = r[n][1].text

    
json_line_attrib('abstract_pt','/article/front/article-meta/abstract[@xml:lang="pt"]')

print('========== autores citados ==================')

r=tree.xpath('//*[@person-group-type="author"]/*')
print(len(r))
for s in range(len(r)):
    print(s)
    print(r[s][0].text, r[s][1].text)
    
print('========== autores citados 2 ==================')
r=tree.xpath('/article/back/ref-list/ref/nlm-citation/person-group[@person-group-type="author"]/*')
print(len(r))
for s in range(len(r)):
    print(s)
    print(r[s][0].text, r[s][1].text)

    
print('===== fontes citadas ============')
r=tree.xpath('/article/back/ref-list/ref/nlm-citation/source')
n=tree.xpath('/article/back/ref-list/ref/nlm-citation/year')
print(len(r))
print(len(n))
for s in range(len(r)):
    print(r[s].text, n[s].text)
    
print('===== artigos citados ============')
r=tree.xpath('/article/back/ref-list/ref/nlm-citation/article-title')
print(len(r))
print(len(n))
for s in range(len(r)):
    print(r[s].text, n[s].text)
    

for key, value in elem.items() :
    print (key,':',value)


```
